```python
# Used to test the cosine similarity implementation created in Rust

from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity

normalize(np.array([[0, 1, 2, 3, 4]]))

x = np.array([0., 1., 2., 3., 4., 5., 6., 7.])
y = np.array([0.5, 1.2, 2., 3., -1., 0., 6., 7.])

divisor = np.sqrt((x * x).sum())
scaled_x = x / divisor

divisor = np.sqrt((y * y).sum())
scaled_y = y / divisor

dot_sum = 0
for i in range(0, len(x)):
    dot_sum += scaled_x[i] * scaled_y[i]
    
print("Manual cosine similarity: ", dot_sum)

print("From sklearn: ", cosine_similarity(np.array([x]), np.array([y])))
```
